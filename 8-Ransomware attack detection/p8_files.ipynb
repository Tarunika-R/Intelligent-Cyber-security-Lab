```python
import os
import random
import string
from reportlab.pdfgen import canvas

# Create output directory
output_dir = "D:\TARU\V th year\Intelligent Cyber security Lab\ex8\output_pdfs"
os.makedirs(output_dir, exist_ok=True)


# Helper function to generate random string names
def random_filename(length=8):
    return "".join(random.choices(string.ascii_lowercase, k=length))


# Helper function to generate random content
def random_content(words=20):
    return " ".join(
        "".join(random.choices(string.ascii_lowercase, k=random.randint(3, 8)))
        for _ in range(words)
    )


# Generate 50 PDF files
for i in range(50):
    filename = random_filename() + ".pdf"
    filepath = os.path.join(output_dir, filename)
    content = random_content()

    # Create PDF with ReportLab
    c = canvas.Canvas(filepath)
    c.drawString(100, 750, f"Random PDF #{i+1}")
    c.drawString(100, 730, f"Filename: {filename}")
    c.drawString(100, 700, "Content:")
    c.drawString(100, 680, content)
    c.save()

    print(f"âœ… Created: {filename}")

print("\nAll 50 PDFs created in 'output_pdfs/'")
```
